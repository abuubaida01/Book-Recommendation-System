# **Popularity Base Recommendation System**

- **I will show top 50 Highest rating books, on which at least 250 people mush have voted.** 

```python
rating.head()
```

```python
book.head()
```

##### **first let's merge book and rating Df on ISBN**


```python
df = book.merge(rating, on="ISBN")
df.head(3)
```

- now I will find the total rating of each book, and will carry just top rating books

```python
# now let's groupby w.r.t title, as ISBN number is unique for same book, so Title is appropriate 
df.groupby('Book-Title').count()['Book-Rating']

#here we got the series, let's convert it into Data Frame
num_rat_df=df.groupby('Book-Title').count()['Book-Rating'].reset_index()

## here Book-Rating is the number of rating give to the single book
num_rat_df.head(2)
```

```python
#let's change the name of Book_rating to no-of-rating
num_rat_df.rename(columns={'Book-Rating':"no-of-rating"}, inplace= True)
```

```python
## let's get the average of total no of rating 
avg_rat_df=df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rat_df.rename(columns={'Book-Rating':"avg-of-rating"}, inplace= True)

avg_rat_df.head(2)
```

##### let's merge both of them on Book-Title


```python
new_df=num_rat_df.merge(avg_rat_df, on='Book-Title')
new_df
```

```python
## now, I will carry just those books where the no_of_rating is greater than 250
top_50 = new_df[new_df['no-of-rating']>=250].sort_values('avg-of-rating', ascending=False).head(50)
```

```python
top_50
```

```python
# I need author-name and pic of these books, so let's merge on top of this DF
top_50.merge(df, on="Book-Title").shape
```

```python
## I think there are some duplicates on book-titles, so let's remove them 
top_50 = top_50.merge(df, on="Book-Title").drop_duplicates('Book-Title')
top_50.shape
```

```python
top_50.columns
```

```python
# now I need just few columns 
top_50=top_50[['Book-Title', 'Book-Author', 'no-of-rating','avg-of-rating','Image-URL-M']]
top_50.head()
```

```python
top_50['avg-of-rating'] = top_50['avg-of-rating'].round(decimals=1)
top_50.head()
```

## Collaborative Recommended system 
- need user which must have voted at least 200 times 
- Need book which must have been rated at least 50 times


```python
rating.head()
# user.head()
```

- **let's take out only those user-ID, who have voted more than 200**  

```python
x = rating.groupby('User-ID')['Book-Rating'].count() > 200 ## we got boolean 
edu_users = x[x].index ## only true will come up 

edu_users.shape
```

```python
user_rater_df = df[df['User-ID'].isin(edu_users)]  

# here we got all those user which fulfil our requirement 
# only 900 people rated :) rest were chilling  
```

- **Now I will take out those book, which have been voted more than 50 times**

```python
y = user_rater_df.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index  # here those book are excluded which have been rated at least than 50 time 
final_df =user_rater_df[user_rater_df['Book-Title'].isin(famous_books)]
final_df.head() # this is fulfilling are requirement 

final_df.drop_duplicates()
```

### **Let's make Pivot Table for getting Similarity**

```python
# now let's make new data frame
pt = final_df.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.head()

pt.fillna(0, inplace=True)
pt.head()
```

---

### **Similarity score** 

```python
## NOw let's find the distance btw each book, here each book will be consider as vector, 
## I will apply cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(pt)
```

```python
cs ## this is the calculated Similarity of each book with others.
```

---

#### **Process of Getting Recommendation**

```python
## Suppose I got the name of book, first I will find the index 
x = np.where(pt.index == '1984')[0][0]

## no  I will go to this 
list(enumerate(cs[x])) # by enumerate function we can get the index number as well
sorted(list(enumerate(cs[x]))) # it is sorted by index, & I need to sort by value
sorted(list(enumerate(cs[x])), key =lambda x: x[1], reverse=True) # sorted in descending order
sorted(list(enumerate(cs[x])), key=lambda x: x[1], reverse=True)[1:6] # now I need just upper 6
```

---

### **Final Function**

```python
# now let's make a function 
def recommend(book_title):
    # book_name = str(book_title)
    index = np.where(pt.index == book_title)[0][0]
    similar_item=sorted(list(enumerate(cs[index])), key=lambda x: x[1], reverse=True)[1:11] 

    data= []
    for i in similar_item: 
        # print(i) # we got index, need index[i[0]]
        ## now I will match this title with each title of Books and will fetch book author name and Image
        # print(pt.index[i[0]])
        item= []
        # here book must bee chosen by top-50 books so that he/she can see recommendation
        book['Book-Title'] = book['Book-Title'].drop_duplicates()

        item.extend(book[book['Book-Title']==pt.index[i[0]]]['Book-Title'])
        item.extend(book[book['Book-Title']==pt.index[i[0]]]['Book-Author'])
        item.extend(book[book['Book-Title']==pt.index[i[0]]]['Image-URL-M'])
        data.append(item)  ## we will get the list of list 
    return data
    
```
